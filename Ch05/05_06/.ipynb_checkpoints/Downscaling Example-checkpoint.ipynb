# Downscaling Example


```python
# Let's work with the picture of a house

import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = 'retina'
from matplotlib import rcParams
from EC_CV import *
import cv2

rcParams['figure.figsize'] = 20,8

img = plt.imread('volcano.jpg')
#plt.axis("off")
plt.imshow(img)
np.shape(img)
```

```python
kernel = np.matrix([[0, -1, -2], 
                    [10,  0, -1],
                    [2,  1,  0]])

kernel = np.array([[[0,0,0], [-1,0,0], [-2,0,0]], 
                    [[10,0,0],  [0,0,0], [-1,0,0]],
                    [[2,0,0],  [1,0,0],  [0,0,0]]])

print(kernel[1:3,0:2,:])
print(np.average(kernel[1:3,0:2,:]))
```

```python
# Let's reduce the resolution by a factor of 3 (9 times smaller)

def downscale_by_3(pic):
    rows, cols, temp = np.shape(pic)
    rows = int(rows/3)
    cols = int(cols/3)
    smaller = np.zeros((rows,cols,3),np.float64)
    for i in range(rows):
        for j in range(cols):
            smaller[i,j,0] = np.average(pic[i*3:i*3+3,j*3:j*3+3,0])  #    Red Channel
            smaller[i,j,1] = np.average(pic[i*3:i*3+3,j*3:j*3+3,1])  # Green Channel
            smaller[i,j,2] = np.average(pic[i*3:i*3+3,j*3:j*3+3,2])  # Blue Channel
    return smaller
        
img_s = adapt_image(downscale_by_3(img))
plt.imshow(img_s)
```

```python
# Let's see the two images side by side

rcParams['figure.figsize'] = 20,8

# display images
fig, ax = plt.subplots(1,2)
ax[0].imshow(img)
ax[1].imshow(img_s)
```

```python
# Now let's do the same at the smaller image's size

rcParams['figure.figsize'] = 5,8

# display images
fig, ax = plt.subplots(1,2)
ax[0].imshow(img)
ax[1].imshow(img_s)
```

```python
# Finally let's save the smaller image into a file

plt.imsave("smaller.jpg",img_s)
```
